```python
# Mean income of different sex
sex_income = {}
for i in range(len(data_dict['Sex'])):
    sex = data_dict['Sex'][i]
    income = int(data_dict['Income'][i])
    if sex not in sex_income:
        sex_income[sex] = []
    sex_income[sex].append(income)

for sex, incomes in sex_income.items():
    avg_income = sum(incomes) / len(incomes)
    print(f'Mean income for Sex {sex}: {avg_income:.2f}')
```

```python
# Mean income of different education
education_income = {}
for i in range(len(data_dict['Education'])):
    education = data_dict['Education'][i]
    income = int(data_dict['Income'][i])
    if education not in education_income:
        education_income[education] = []
    education_income[education].append(income)

for education, incomes in education_income.items():
    avg_income = sum(incomes) / len(incomes)
    print(f'Mean income for {education}: {avg_income:.2f}')
```

```python
# Mean income of different settlement size
settlement_income = {}
for i in range(len(data_dict['Settlement size'])):
    settlement = data_dict['Settlement size'][i]
    income = int(data_dict['Income'][i])
    if settlement not in settlement_income:
        settlement_income[settlement] = []
    settlement_income[settlement].append(income)

for settlement, incomes in settlement_income.items():
    avg_income = sum(incomes) / len(incomes)
    print(f'Mean income for Settlement size {settlement}: {avg_income:.2f}')
```

```python
#rate of education with different settlement size
education_by_settlement = {}

for i in range(len(data_dict['Settlement size'])):
    settlement = data_dict['Settlement size'][i]
    education = data_dict['Education'][i]

    if settlement not in education_by_settlement:
        education_by_settlement[settlement] = {}

    if education not in education_by_settlement[settlement]:
        education_by_settlement[settlement][education] = 0

    education_by_settlement[settlement][education] += 1

# Calculate rates
for settlement, education_counts in education_by_settlement.items():
    total_in_settlement = sum(education_counts.values())
    print(f'Education distribution in Settlement size {settlement}:')
    for education, count in education_counts.items():
        rate = (count / total_in_settlement) * 100
        print(f'  {education}: {count} ({rate:.2f}%)')
```

```python
#rate of occupation with different education
occupation_by_education = {}

for i in range(len(data_dict['Education'])):
    education = data_dict['Education'][i]
    occupation = data_dict['Occupation'][i]

    if education not in occupation_by_education:
        occupation_by_education[education] = {}

    if occupation not in occupation_by_education[education]:
        occupation_by_education[education][occupation] = 0

    occupation_by_education[education][occupation] += 1

# Calculate rates
for education, occupation_counts in occupation_by_education.items():
    total_in_education = sum(occupation_counts.values())
    print(f'Occupation distribution in education size {education}:')
    for occupation, count in occupation_counts.items():
        rate = (count / total_in_education) * 100
        print(f'  {occupation}: {count} ({rate:.2f}%)')
```

```python
import matplotlib.pyplot as plt
import numpy as np

occupation = data_dict['Occupation']
education = data_dict['Education']
settlement = data_dict['Settlement size']

# Map occupation to 'Skilled' or 'Unskilled'
def occ_group(o):
    if o == 'skilled employee / official' or o == 'management / self-employed / highly qualified employee / officer':
        return 'Skilled'
    else:
        return 'Unskilled'

# Group counts: (occupation_group, education, settlement) -> count
group_counts = {}
for i in range(len(occupation)):
    o = occ_group(occupation[i])
    e = education[i]
    s = settlement[i]
    key = (o, e, s)
    if key not in group_counts:
        group_counts[key] = 0
    group_counts[key] += 1

# Prepare unique values
education_types = sorted(list(set(education)))
settlement_types = sorted(list(set(settlement)))
occ_types = ['Skilled', 'Unskilled']

# Plot as 2x2 subplots (one for each education level)
fig, axes = plt.subplots(2, 2, figsize=(16, 10))
axes = axes.flatten()

for idx, edu in enumerate(education_types):
    skilled_counts = []
    unskilled_counts = []
    for s in settlement_types:
        skilled_counts.append(group_counts.get(('Skilled', edu, s), 0))
        unskilled_counts.append(group_counts.get(('Unskilled', edu, s), 0))
    x = np.arange(len(settlement_types))
    ax = axes[idx]
    ax.bar(x - 0.2, skilled_counts, 0.4, label='Skilled')
    ax.bar(x + 0.2, unskilled_counts, 0.4, label='Unskilled')
    ax.set_title(f'Education: {edu}')
    ax.set_xticks(x)
    ax.set_xticklabels(settlement_types)
    ax.set_ylabel('Count')
    ax.legend()

plt.suptitle('Occupation (Skilled vs Unskilled) × Education & Settlement Size')
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.savefig('occupation_vs_education_settlement_1.png')
plt.show()
```
