```python
import statistics
file = open('/Users/henryhenderson/Documents/Python/student_habits_performance.csv', 'r')
lines = file.readlines()
file.close()

header = lines[0].strip().split(',')

data_dict = {col: [] for col in header}

for line in lines[1:]:
    row = line.strip().split(',')

    if len(row) == len(header):
        for i, col in enumerate(header):
            value = row[i] if row[i] != "" else None
            data_dict[col].append(value)
```

```python
import statistics
import math

def corr(x, y):
    mean_x = statistics.mean(x)
    mean_y = statistics.mean(y)
    num = sum((a - mean_x) * (b - mean_y) for a, b in zip(x, y))
    den = math.sqrt(sum((a - mean_x) ** 2 for a in x) * sum((b - mean_y) ** 2 for b in y))
    return round(num / den, 2)

print("\nRelationships:")
print("Sleep vs Mental Health correlation:", corr(sleep_hours, mental_health))
print("Social Media vs Exam Score correlation:", corr(social_media, exam_scores))
print("Study Hours vs Exam Score correlation:", corr(study_hours, exam_scores))
print("Exercise vs Mental Health correlation:", corr(exercise, mental_health))
# Correlations
```

```python
# Grouped aggregate: Part-time job vs Study Hours
job_study_hours = {"part_time": [], "no_job": []}

for i, v in enumerate(data_dict["study_hours_per_day"]):
    if v not in (None, "", "None"):
        try:
            hours = float(v)
            job_status = data_dict["part_time_job"][i]

            # Normalise job_status (convert everything to lowercase string)
            job_status_str = str(job_status).strip().lower()

            if job_status_str in ("1", "yes", "true"):
                job_study_hours["part_time"].append(hours)
            elif job_status_str in ("0", "no", "false"):
                job_study_hours["no_job"].append(hours)

        except:
            continue

import statistics

print("\nStudy Hours by Part-Time Job Status")
for group, vals in job_study_hours.items():
    if vals:
        mean = round(statistics.mean(vals), 2)
        median = statistics.median(vals)
        stdev = round(statistics.stdev(vals), 2) if len(vals) > 1 else 0
        print(f"{group}: n = {len(vals)}, mean = {mean}, median = {median}, stdev = {stdev}")
```
